```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
TrainStatsReader is used to read training stats and draw figures.
"""
class TrainStatsReader:
    
    def __init__(self, train_stats_path):
        stats, reward = np.load(train_stats_path, allow_pickle=True)
        self.stats = stats
        self.reward = reward
        self.iter_num = len(reward)
    
    
        
```

```python
train_stats = TrainStatsReader('20200504_stats_rewards_G_medium_size.npy')
reward = train_stats.reward
iter_num = len(reward)
our_agent_reward = [reward[i][1] for i in range(iter_num)]
random_agent_reward = [reward[i][2] for i in range(iter_num)]
iter_list = list(range(iter_num))
plt.plot(iter_list, our_agent_reward)
plt.plot(iter_list, random_agent_reward)
plt.xlabel('iteration', fontsize=15)
plt.ylabel('reward', fontsize=15)
plt.legend(['our agent', 'random agent'])
plt.savefig('./figs/train_rewards.pdf', format='pdf')
```
